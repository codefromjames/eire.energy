```python
uploaded = files.upload()
ppr = pd.read_csv(io.BytesIO(uploaded['PPR-ALL.csv']), encoding= 'unicode_escape')
# Dataset is now stored in a Pandas Dataframe

```

```python
ppr['index'] = np.arange(len(ppr))
ppr['AddressCO'] = ppr['Address'] + ', ' +ppr['County']  
ppr
```

```python
from geopandas.tools import geocode
from google.colab import files

import time

for index, row in ppr.iterrows():
  try: 
      time.sleep(1)    
      print(row['AddressCO'])
      info = geocode(str(row['AddressCO']), provider = 'nominatim', user_agent = 'my_request' )
      ppr.loc[int(index), 'found'] = info['address'].loc[0]
      ppr.loc[int(index), 'Lon'] = info['geometry'].loc[0].x
      ppr.loc[int(index), 'Lat'] = info['geometry'].loc[0].y
      print(row['index'])
  except Exception as e:
      ppr.loc[int(index), 'found'] =  "No Result"


ppr.to_csv('ppr.geo.csv') 
files.download('ppr.geo.csv')
```
